# 05_search_sample

参考文献文字列をトークン化し、Solr に候補検索を投げて上位結果を確認する。

```python
# Notebook から src 配下の自作モジュールを import できるようにする設定
from pathlib import Path
import sys

project_root = Path.cwd().parent
src_path = project_root / "src"
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))
```

```python
# 設定と検索関数を読み込む
from pprint import pprint

from tanigawa_shoshi.config import SEARCH_FIELD, SOLR_BASE_URL, SOLR_CORE
from tanigawa_shoshi.search import build_query_tokens, search_reference, split_reference_values
```

```python
# 参考文献文字列の入力
reference_text = "畠山 久尚：地滋氣及び地電流の灣型變化並に脈動に關する研究, 気象集誌. 第2輯, Vol. 17, No. 6, pp. 209-229, 1939"
rows = 10

print(reference_text)
```

```python
# 検索前に分割結果と token を確認する
reference_values = split_reference_values(reference_text)
query_tokens = build_query_tokens(reference_text)

print("reference values")
print(reference_values)
print()
print("query tokens")
print(query_tokens)
print()
print("search field")
print(SEARCH_FIELD)
```

```python
# Solr に検索を投げる
search_result = search_reference(
    reference_text,
    rows=rows,
    solr_base_url=SOLR_BASE_URL,
    core_name=SOLR_CORE,
)

print("search params")
pprint(search_result["params"])
print()
print(f"hit count: {search_result['results'].hits}")
```

```python
# 上位結果を確認する
for index, doc in enumerate(search_result["results"], start=1):
    print("=" * 60)
    print(f"rank: {index}")
    print(f"score: {doc.get('score')}")
    print(f"doi: {doc.get('doi')}")
    print(f"title: {doc.get('title')}")
    print(f"authors: {doc.get('authors')}")
    print(f"journal: {doc.get('journal')}")
    print(f"year: {doc.get('year')}")
    print(f"volume: {doc.get('volume')}")
    print(f"page: {doc.get('page')}")
```
