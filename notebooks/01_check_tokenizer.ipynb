```python
# 自作モジュールのインポート
from tanigawa_shoshi.tokenizer import build_segments, split_units, tokenize, tokenize_values
```

#### ICU分割、セグメント生成([{type, item}, {type, item}...])、トークナイズ の実行チェック

```python
sample = "畠山 久尚：地滋氣及び地電流の灣型變化並に脈動に關する研究, 気象集誌. 第2輯, Vol. 17, No. 6, pp. 209-229, 1939"
units = split_units(sample)
segments = build_segments(units)
tokens = tokenize(sample)

print("units:")
print(units)

print("\nsegments:")
print("[")
for segment in segments:
    print("  {")
    print(f"    type: {segment['type']},")
    print(f"    items: {segment['items']}")
    print("  },")
print("]")

print("\ntokens:")
print(tokens)
```
