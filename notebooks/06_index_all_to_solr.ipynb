# 06_index_all_to_solr

MongoDB の JaLC 文書を全件読み込み、Solr へ batch 登録する。

```python
# Notebook から src 配下の自作モジュールを import できるようにする設定
from pathlib import Path
import sys

project_root = Path.cwd().parent
src_path = project_root / "src"
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))
```

```python
# 設定と全件登録関数を読み込む
from pprint import pprint

from tanigawa_shoshi.config import BATCH_SIZE, SOLR_BASE_URL, SOLR_CORE
from tanigawa_shoshi.solr_indexer import delete_all_documents, get_solr_client, index_all
```

```python
# 実行設定
# full 実行時は limit=None のまま使う
limit = None
batch_size = BATCH_SIZE

print(f"core: {SOLR_CORE}")
print(f"batch_size: {batch_size}")
print(f"limit: {limit}")
```

```python
# 全件登録を実行する
# 実際に登録する時だけこのセルを実行する
stats = index_all(
    solr_base_url=SOLR_BASE_URL,
    core_name=SOLR_CORE,
    batch_size=batch_size,
    limit=limit,
)

pprint(stats)
```

```python
# Solr コアは残したまま、登録済み文書だけを全削除する
# 全件入れ直しの前に空に戻したい時だけこのセルを実行する
solr = get_solr_client(SOLR_BASE_URL, SOLR_CORE)
delete_stats = delete_all_documents(solr)

print("delete stats")
pprint(delete_stats)
```
