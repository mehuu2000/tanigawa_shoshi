```python
# Notebook から src 配下の自作モジュールを import できるようにする設定
from pathlib import Path
import sys

project_root = Path.cwd().parent
src_path = project_root / "src"
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))
```

```python
# 設定と schema 操作関数のインポート
from pprint import pprint

from tanigawa_shoshi.config import SOLR_BASE_URL, SOLR_CORE
from tanigawa_shoshi.solr_schema import ensure_schema, get_existing_fields, get_expected_fields
```

```python
# 期待しているフィールド定義を確認する
for field in get_expected_fields():
    pprint(field)
```

```python
# schema を確認し、不足フィールドを追加する
result = ensure_schema(SOLR_BASE_URL, SOLR_CORE)
pprint(result)
```

```python
# 今回使うフィールドだけを見やすく確認する
existing_fields = get_existing_fields(SOLR_BASE_URL, SOLR_CORE)
for field in get_expected_fields():
    print("=" * 60)
    pprint(existing_fields.get(field["name"]))
```
